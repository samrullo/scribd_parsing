```python
import time

start_time = time.time()
time.sleep(3)
end_time = time.time()
print(f"{int(end_time-start_time)} seconds passed")
```

```python
def ask_to_stop():
    answer = input("Do you want to stop?")
    if answer.lower() in ["y","yes"]:
        return True
    else:
        return False
```

```python
should_stop = ask_to_stop()
print(should_stop)
```

```python
def extract_the_whole_page_text():
    text_line_elements = driver.find_elements(By.CSS_SELECTOR,".text_line")
    logging.debug(f"found {len(text_line_elements)} text_line elements")
    text = " ".join([element.text for element in text_line_elements])
    return text
```

```python
def get_next_button():
    return driver.find_element(By.CSS_SELECTOR,".page_right.next_btn")
```

```python
def did_reach_last_page(page_counter,number_of_pages,pages_per_view=2):
    effective_num_of_pages = number_of_pages//pages_per_view
    return page_counter>=effective_num_of_pages
```

```python
wait_time = 3
start_time=time.time()
counter=1
should_stop = False
number_of_pages = 650

while not should_stop:
    try:
        logging.info(f"STARTING {counter} iteration")
        page_text = extract_the_whole_page_text()
        file = folder/f"page_{counter}.txt"
        file.write_text(page_text, encoding="utf-8")
        next_button = get_next_button()
        next_button.click()

        end_time = time.time()
        passed_time = int(end_time - start_time)
        if passed_time > 60 * 60:
            should_stop = ask_to_stop()
        if did_reach_last_page(counter,number_of_pages):
            should_stop = ask_to_stop()
        time.sleep(wait_time)
    except Exception as e:
        print(e)
        should_stop = ask_to_stop()
    counter += 1
```
